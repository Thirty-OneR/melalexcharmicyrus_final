```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_json("data/2020oi_data.json")
```

```python
size = len(data['2020oi']['photometry'][0]['data'])
mag = []
err = []
bands = []
obs_date = []
time = []
for x in np.arange(size):
    temp = data['2020oi']['photometry'][0]['data'][x]
    obs_date.append(data['2020oi']['photometry'][0]['data'][x]['fields']['obs_date'])
    mag.append(data['2020oi']['photometry'][0]['data'][x]['fields']['mag'])
    err.append(data['2020oi']['photometry'][0]['data'][x]['fields']['mag_err'])
    bands.append(data['2020oi']['photometry'][0]['data'][x]['fields']['band'])
```

```python
from datetime import datetime
from dateutil import parser
import time 

timearr = []
dates = []
for date in obs_date: 
    date = date.replace("Z","")
    date = date.replace("T"," ")
    betterdate = parser.parse(date)#, '%Y-%m-%d %I:%M%p')
    dates.append(betterdate)
    timearr.append(time.mktime(betterdate.timetuple()))
```

```python
optical_df = pd.DataFrame({'time':timearr, 'mag':mag, 'err':err, 'band':bands})
```

```python
optical_df.to_csv("./data/2020oi_data.csv", index=False)
```

```python
optical_df['lamb(nm)'] = np.nan
optical_df.loc[optical_df['band'] == 'ZTF-Cam - r-ZTF', 'lamb(nm)'] = 658
optical_df.loc[optical_df['band'] == 'ZTF-Cam - g-ZTF', 'lamb(nm)'] = 475
```

```python
sns.scatterplot(x='time', y='mag', hue='lamb(nm)',data=optical_df);
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

# Create plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)
ax = fig.gca(projection='3d')
ax.view_init(azim=80, elev=20)
ax.invert_zaxis()
ax.invert_xaxis()
ax.scatter(optical_df['time'].values, optical_df['lamb(nm)'].values, optical_df['mag'].values, c=optical_df['lamb(nm)'], cmap='Dark2', alpha=0.8,s=30)
ax.set_xlabel("Time (s)",fontsize=16)
ax.set_ylabel(r"$\lambda$ (nm)",fontsize=16)
ax.set_zlabel("mag",fontsize=16)
plt.show()
```

```python
#fit quadratic as a function of time 
oneBand = optical_df.loc[optical_df['lamb(nm)'] == 475]
plt.plot(oneBand['time'], oneBand['mag'], 'o')
plt.gca().invert_yaxis()
```

```python
coefs_time = np.polyfit(oneBand['time'], oneBand['mag'], 2)
```

```python
plt.plot(oneBand['time'], coefs_time[0]*oneBand['time']**2 + coefs_time[1]*oneBand['time'] + coefs_time[2], c='r', label='fit')
plt.plot(oneBand['time'], oneBand['mag'], 'o', label='data')
plt.gca().invert_yaxis()
plt.legend();
```
